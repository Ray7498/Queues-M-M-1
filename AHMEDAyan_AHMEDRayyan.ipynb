Let us start by defining what a Markov Jump Process essentially is. A Markov Jump Process is basically a Markov Process in which we wait in a particular state for a random amount of **'holding time'** and then we suddenly jump to another state. We can break this process into two seperate ideas:
- The first thing that comes to our mind would be where we are going to jump to. This is respresented by a *jump chain* which is a discrete time Markov Chain telling us exactly what transitions are made but it doesn't tell us when those transitions are happening.
- Now that we know where we are going to jump, it is also necessary to know how long we have to wait before jumping. Knowing that we will also get to know when those jumps are going to happen. This idea is represented by *holding times*. Since we are emphasizing on the word *Markov* in these processes, they should also satisfy the Markov property. This brings us to the fact that these holding times follow an Exponential Distribution. But why an Exponential Distribution? That is because it is the only distribution which follows the memoryless property.

**Theorem 1.1:** *Let $K~\sim \text{Exp}(\lambda)$. Then for any $k_1,k_2\geq 0$,$$\mathbb{P}(K>k_1+k_2|K>k_1)=\mathbb{P}(K>k_2)$$*

This property of an Exponential Distribution is known as the *memoryless property.*

*Proof.*
By standard use of conditional probability, we have
$$
\mathbb{P}(K > k_{1} + k_{2} \,|\, K > k_{1}) = \frac{\mathbb{P}(K > k_{1} + k_{2} \cap K > k_{1})}{\mathbb{P}(K > k_{1})} = \frac{\mathbb{P}(K > k_{1} + k_{2})}{\mathbb{P}(K > k_{1})} = \frac{e^{-\lambda(k_{1}+k_{2})}}{e^{-\lambda k_{1}}} = e^{-\lambda k_{2}},
$$
which is still the tail probability of the exponential distribution.

$$$$


Let us consider a Markov Jump process $(N_t)$ on a state space $S$. We can represent the **transition rate** to jump from state $i$ to state $j$ where $i,j \in S, i\neq j$ by $q_{ij}\geq 0$. This means that after a time with an $\text{Exp}(q_{ij})$, the process jumps to state $j$ if it has not jumped yet.

So from state $i$, there are plenty other states $j$ that we can jump to each with a waiting time $\text{Exp}(q_{ij})$. We already know that the distribution of the minimum of the $\text{Exp}(q_{ij})$'s is itself an Exponential distribution with parameter: $$q_i = \sum_{j\neq i}q_{ij}$$

Also, the probability that we move from some state $i$ to $j$ is given by: $$r_{ij}=\frac{q_{ij}}{\sum_{j\neq i}q_{ij}}=\frac{q_{ij}}{q_i}$$

So, starting from state $i$, we wait for an exponential amount of time Exp($q_i$) and then move to a state $j$ with probability $r_{ij}$. So we can define the transition rates in the form of a **Generator Matrix $Q$** such that the off-diagonal entries are positive and equal to $q_{ij}$ where as the diagonal entries are negative or 0 and equal to $$q_{ii}=-\sum_{j\neq i}q_{ij}$$ Thus the sum of entries in each row is equal to 0. Using this, we can easily define the transition matrix $R$ with the help of $r_{ij}$ defined above.

Now, to keep track of the movement of the process, we can consider the **discrete time jump Markov chain** $X_{n}$ associated with ($N_t$) as:

$$X_0=N_0$$ and $$X_n= \text{state of}\; N_t\; \text{just after}\; n_{th}\; \text{jump}$$

So, once we know where the jump process will move, we will get to know the next holding time from state $j$ which will be Exp($q_j$) and the process goes on.

We can summarize all this into a formal definition as follows:

**Definition 1.1:** 
* Let $S$ be a set and $\lambda$ a distribution on $S$.
* Let $Q=(q_{ij}:i,j\in S)$ be a matrix where $q_{ij}\geq 0$ for $i\neq j$ and $\sum_j q_{ij}=0$ for all $i$, and write $q_{ii}=-q_i=- \sum_{j\neq i}q_{ij}$.
* Define  $R=(r_{ij}:i,j\in S)$ be a matrix as follows: for $i$ such that $q_{i}\neq 0,$ write $r_{ij}= \frac{q_{ij}}{q_i}$ for $j\neq i$ and $r_{ii}=0$ and for $i$ such that $q_{i}=0$, write $r_{ij}=0$ for $j\neq i$ and $r_{ii}=1$.

We define the **Markov Jump Process** on the state space $S$ and generator matrix $Q$ and initial distribution $\lambda$ as follows:

* The **jump chain** ($X_{n}$) is the discrete time Markov Chain on $S$ with initial distribution $\lambda$ and transition matrix $R$.
* The **holding times** $T_{1}-T_{0}, T_{2}-T_{1},\ldots $have distribution $T_{n}-T_{n-1}\sim \text{Exp}(q_{X_{n-1}})$, and are conditionally independent given $X_{n}$.
* The **jump times** are $T_{n}$.

Then the **Markov Jump Process $(N_{t})$** is defined by:
$$
N_{t} = \begin{cases} 
X_0 & \text{for } t < T_1 \\
X_n & \text{for } T_n \leq t < T_{n+1}
\end{cases}
$$

#### Transition Matrix

For a continuous-time Markov chain, we define the **transition matrix** $ P(t) $. The $ (i, j) $th entry of the transition matrix is given by

$$ P_{ij}(t) = \mathbb{P}(N_t = j | N_0 = i). $$

The transition matrix satisfies the following properties:

1. $ P(0) $ is equal to the identity matrix, $ P(0) = I $;
2. The rows of the transition matrix must sum to 1,

$$ \sum_{j \in S} P_{ij}(t) = 1, \quad \text{for all} \ t \geq 0; $$

3. For all $ s, t \geq 0 $, we have

$$ P(s + t) = P(s)P(t). $$

Moreover, alternatively, we can also define a Markov Jump Process as mentioned in the lecture notes as follows:

**Definition 1.2.** 
A jump process is a random variable $ X : (\Omega, \mathcal{A}, P) \rightarrow (X, \mathcal{T}) $, where

- $ \mathcal{A} $ is a $ \sigma $ field on $ \Omega $ and $ P $ a probability measure on $ (\Omega, \mathcal{A}) $
- The space $ X $ is the set of functions from $ [0, \infty) $ to $ E $, which are continuous on the right and limited on the left (cadlag):

$$ X = \{ f : [0, \infty) \rightarrow E, \quad \lim_{s \to t-, s < t} f(s) = f(t-), \quad \lim_{s \to t, s > t} f(s) = f(t) \} $$

- The $ \sigma $ field $ \mathcal{T} $ is generated by marginals, i.e. it is the smallest $ \sigma $ field of $ X $ such that if $ f \in X \rightarrow f(t) $ is measurable for any $ t \geq 0 $.

**Theorem 1.2** Let $ (N_t) $ be a Markov jump process on a state with generator matrix $ Q $. If $ \pi = (\pi_i) $ is a distribution with $ \sum_{i}\pi_{i}q_{ij} = 0 $ for all $ j $, then $ \pi $ is a stationary distribution. In matrix form, this condition is $ \pi Q = 0 $.

*Proof.* Suppose $ \pi Q = 0 $. We need to show that $ \pi P(t) = \pi $ for all $ t $. We'll first show that $ \pi P(t) $ is constant, then show that constant is indeed $ \pi $.

First, we can show that $ \pi P(t) $ is constant by showing that its derivative is zero. Using the backward equation $ P'(t) = QP(t) $, we have

$$
\frac{d}{dt} \pi P(t) = \frac{d}{dt} (\pi QP(t)) = \pi QP'(t) = 0P(t) = 0,
$$

as desired.

Second, we can find that constant value by setting $ t $ to any value: we pick $ t = 0 $. Since $ P(0) = I $, the identity matrix, we have

$$
\pi P(t) = \pi P(0) = \pi
$$

for all $ t $.                                                        $\square$


We now write,

$$
V_j(t) = \int_0^t \mathbb{1}_{N_s = j} ds
$$

for the total amount of time spent in state $ j $ up to time $ t $. Here, the indicator function $ \mathbb{1}_{N_s = j} $ is 1 when $ N_s = j $ and 0 otherwise, so the integral is measuring the time we spend at $ j $. So $ V_j(t)/t $ is the proportion of time up to time $ t $ spent in state $ j $, and we interpret its limiting value (if it exists) to be the long-run proportion of time spent in state $ j $.


**Theorem 1.3 (Ergodic theorem)** Let $ (N_t) $ be an irreducible Markov jump process with generator matrix $ Q $. Then for any initial distribution $ \lambda $ we have, for that $ \forall j, \frac{V_j(t)}{t} \to \frac{1}{q_{j}\mu_j} $ almost surely as $ t \to \infty $, where $ \mu_j $ is the expected return time to state $ j $ unless the state space is a single absorbing state $ j $, in which case $ V_j(t) \to 1 $ almost surely.

* Suppose $ (N_t) $ is positive recurrent. Then there is a unique stationary distribution $ \pi $ given by $ \pi_j = \frac{1}{q_{j}\mu_j} $ (unless the state space is a single absorbing state $ j $, in which case $ \pi_j = 1 $), and $ \frac{V_j(t)}{t} \to \pi_j $ almost surely for all $ j $.
    
* Suppose $ (N_t)$ is null recurrent or transient. Then $ \frac{V_j(t)}{t} \to 0 $ almost surely for all $ j $.

We can also state the Ergodic Theorem as follows:

***Theorem 1.4 (Ergodic Theorem)** Let $(N_t)$ be an irreducible and positive recurrent Markov Jump Process on state space $E$. Let $\pi$ be its invariant probability and $f: E\rightarrow \mathbb{R}$ bounded. Then, $$ \frac{1}{T}\int_{0}^{T}f(N_t)dt \underset{T\rightarrow \infty}{\xrightarrow{a.s}} \int_E f(x)\pi dx $$

or for simulation, we will use, $$ \lim_{T\rightarrow \infty}\frac{1}{T}\int_{0}^{T}f(N_t)dt=\pi(f) $$

# Queues
There are numerous Markov Jump Processes namely, Poisson processes which we have already studied, Levy Processes, Birth-Death processes, Hawkes Processes and so on.
Here we look at another type of Markov Jump Process, that is Queues. Essentially queues are a specific type of birth-death processes where the birth rate is replaced by the arrival rate and the death rate is replaced by the service rate.

With the definition of Markov Jump Processes in hand, we can come to a more well defined Markov Jump Process, **Queues**.
If we look at the mathematical model of a queue, we come to realise some key features of a queue:
* Firsty, individuals arrive in a queue.
* Then they receive a service.
* Then they leave.

But queues may differ depending on the number of servers available. For instance, if we talk about an Instagram Live Video, it can accomodate infinite users at once and all of them get to receive the service (here, watch the celebrity), at once without actually waiting in a queue. Secondly, if we talk about a professor who is solving students queries, then he can only solve one query at a time and students have to wait in a queue for their turn. Thirdly, while travelling on a highway there are numerous toll stations and the person driving the vehicle can get in either lane to pay the toll fee. Thus, multiple servers are availble at once but they are finite.

In simple terms, we can broadly categorize the queues mentioned above into three simple categories:
* The first case falls under an $M/M/\infty$ queue. The process is referred to as an $M/M/\infty$ process.
* The second case falls under an $M/M/1$ queue. The process is referred to as an $M/M/1$ process.
* The third case falls under an $M/M/c$ queue. The process is referred to as an $M/M/c$ process.

All the above processes are Markov Jump Processes.


## M/M/$\infty$ Process (Brief)
Let us assume that in a given process, the individuals are arriving like a Poisson process of intensity(rate) $\lambda$. On arrival, they receive a service for an $\text{Exp}(\mu)$ amount of time independent of everything else and then they leave. Then the number of individuals $N_{t}$ in the system that are receiving the service ar time $t$ is a Markov Jump process with arrival rates $q_{i,i+1}=\lambda$. Similary, because of Theorem 1.1, the rate of departures when $i$ individuals are being serviced has the rate $q_{i,i-1}=i\mu$ since the first service time to finish is the minimum of $i$ remaining exponential times which is thus exponential with the rate $i\mu$. Thus, using the properties of the generator matrix $Q$, it can be represented as:$$
Q = \begin{pmatrix}
-\lambda & \lambda & 0 & 0 & \cdots \\
\mu & -(\lambda + \mu) & \lambda & 0 & \cdots \\
0 & 2\mu & -(\lambda + 2\mu) & \lambda & \cdots \\
0 & 0 & 3\mu & -(\lambda + 3\mu) & \cdots \\
\vdots & \vdots & \vdots & \vdots & \ddots \\
\end{pmatrix}
$$
The transition rate diagram of this process can be represented as:

This process is called an **$M/M/\infty$** process where, the first M represents the **memoryless arrivals** since arrivals take place as a Poisson process which has exponential interarrival times. The second M represents the **memoryless service times**, since the service times are exponentially distributed with parameter $\mu$. The $\infty$ represents the infinite number of servers available as mentioned in the Instagram Live Video model.

Let us now emphasize on the stationary distribution of the process to find out its long run behaviour. We already know that the stationary distribution will solve the system $\pi Q=0$. Using this, we get: $$
\lambda \pi_{i-1} - (\lambda + i\mu)\pi_i + (i+1)\mu\pi_{i+1} = 0     \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;        (1)
$$
for $i\geq 1$ and $$
-\lambda \pi_0 + \mu \pi_1 = 0                                        \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(2)                          
$$
using the first column of $Q$.
Thus, $\pi_1=\frac{\lambda}{\mu}\pi_0=\rho\pi_0,$ where $\rho=\frac{\lambda}{\mu}$. Now, putting this in the first equation and solving by induction, we get $$
\pi_i = \pi_0 \rho^i / i! = e^{-\rho} \rho^i / i
$$
Here we see that, $\pi_0=e^{-\rho}$ which can be proved by using the normalising equation, i.e, $\sum_i \pi_i=1$. We also infer that the stationary distribution is nothing but the Poisson distribution with parameter $\rho$.

**Theorem 1.5** Let $ (N_{t})$ be an $M/M/$ process with arrivals at rate $\lambda$ and service at rate $\mu$. The process has the stationary distribution $\pi_i = e^{-\rho} \frac{\rho^i}{ i!} $, corresponding to $N_{t} \sim Pois(\rho)$, where $\rho = \frac{\lambda}{\mu} $.

*Proof.* We know that this $ \pi $ is a distribution. We must check that $\pi $ fulfills $ \pi Q = 0 $. We've already checked (2). For $i \geq 1 $, we have from (1),

$$
\begin{align*}
\lambda\pi_{i-1} - (\lambda + \mu)\pi_i + (i + 1)\mu\pi_{i+1} &= \lambda e^{-\rho} \frac{\rho^{i-1}}{(i - 1)!} - (\lambda + \mu) e^{-\rho} \frac{\rho^i}{i!} + (i + 1)\mu e^{-\rho} \frac{\rho^{i+1}}{(i + 1)!} \\
&= \lambda e^{-\rho} \frac{\rho^{i-1}}{(i - 1)!} - (\lambda + \mu) e^{-\rho} \frac{\rho^i}{i!} + \mu e^{-\rho} \rho \frac{\rho^i}{i!} \\
&= e^{-\rho} \frac{\rho^i}{i!} (\lambda \rho - (\lambda + \mu) + \mu \rho) \\
&= e^{-\rho} \frac{\rho^i}{i!} (\lambda \rho - \lambda - \mu + \mu \rho) \\
&= e^{-\rho} \frac{\rho^i}{i!} (\lambda (\rho - 1) - \mu (\rho - 1)) \\
&= e^{-\rho} \frac{\rho^i}{i!} (\rho - 1)(\lambda - \mu) \\
&= 0,
\end{align*}
$$

as desired.

Thus we see that this process indeed has a stationary distribution and simultaneously it is irreducible. Thus, it must be positive recurrent and $\pi$ is also the limiting or equilibrium distribution. So we can use the ergodic theorem to say that the long-run amount of time spent in each state is governed by $\pi$. 

**Theoretical Result 1.1** The average number of individuals being serviced is $\mathbb{E}(Pois(\rho))=\rho=\frac{\lambda}{\mu}$.

**Theoretical Result 1.2** The long-run proportion of time the process is empty is $
\pi_0 = e^{-\rho} = e^{-\lambda/\mu}.
$

## Simulation 

Below is a simplified algorithm to simulate an $M/M/\infty$ process followed by the python code:



1. **Initialize Parameters:**
   - Set the arrival rate ($\lambda_{\text{arrival}}$).
   - Set the service rate ($\lambda_{\text{service}}$).
   - Set the time horizon $(T)$.

2. **Initialize Variables:**
   - Set the current time $t$ to 0.
   - Set the number of customers in the system $(\text{num_customers})$ to 0.
   - Initialize an empty list $\text{events}$  to store both arrival and departure events as tuples $(\text{time}, \text{event_type})$, where $\text{event_type}$ is 1 for arrival and -1 for departure.
   - Initialize an empty list $\text{customer_counts} $ to store the number of customers at each event time.

3. **Simulation Loop:**
   - While $t$ is less than $T$:
     - Generate a random interarrival time $(X)$ from an exponential distribution with rate parameter $\lambda_{\text{arrival}}=\lambda$.
     - Generate a random service time $(Y)$ from an exponential distribution with rate parameter $\lambda_{\text{service}}=\mu$.
     - Update $t$ by adding the minimum of $X$ and $Y$ to it.
     - If $(X < Y)$:
       - Increment $\text{num_customers}$ by 1.
       - Append the tuple $(t, 1)$ to the list $\text{events} $ to indicate an arrival event.
     - Otherwise:
       - If $\text{num_customers}$ is greater than 0:
         - Decrement $\text{num_customers}$ by 1.
         - Append the tuple $(t, -1)$ to the list $ \text{events} $ to indicate a departure event.

4. **Sort Events:**
   - Sort the list $ \text{events} $ based on the time $(\text{time})$.

5. **Update Number of Customers:**
   - Initialize $ \text{current_customers} $ to 0.
   - Iterate through the sorted events:
     - Update $ \text{current_customers} $ by adding the event type $(\text{event_type})$ to it.
     - Append the tuple $(\text{time}, \max(\text{current_customers}, 0))$ to the list $ \text{customer_counts} $ to store the number of customers at each event time. Ensure the number of customers remains non-negative by taking the maximum of $ \text{current_customers} $ and 0.

6. **Plotting:**
   - Plot the step function using the list $ \text{customer_counts} $.
   - Label the x-axis as 'Time'.
   - Label the y-axis as 'Number of customers in the system'.
   - Set the title of the plot as 'M/M/ Markov Jump Process (Queue)'.

7. **Display the Plot:**
   - Show the plot to visualize the number of customers in the system over time.

This algorithm generates arrival and departure events for an $M/M/$ queue until the specified time horizon $ T $. The resulting list of events contains tuples of the form $ (t, \text{num_customers}) $, where $ t $ is the time of the event and $ \text{num_customers} $ is the number of customers in the system at that time.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lambda_arrival = 0.6  # Arrival rate
lambda_service = 0.9  # Service rate
T = 100              # Time horizon

# Simulate M/M/infinity queue
t = 0
num_customers = 0  # Start with zero customers in the system
events = []         # Combined list to store both arrival and departure events

while t < T:
    # Generate time until the next event
    interarrival_time = np.random.exponential(1 / lambda_arrival)
    service_time = np.random.exponential(1 / lambda_service)
    
    # Arrival event
    if interarrival_time < service_time:
        t += interarrival_time
        num_customers += 1
        events.append((t, 1))  # Store arrival event as tuple (time, 1)
    # Departure event
    else:
        t += service_time
        if num_customers > 0:  # Ensure there are customers to depart
            num_customers -= 1
            events.append((t, -1)) # Store departure event as tuple (time, -1)

# Sort events based on time
events.sort(key=lambda x: x[0])

# Initialize variables for plotting
current_customers = 0
customer_counts = [(0, 0)]  # Starting with 0 customers at time 0

# Iterate through events and update number of customers
for event in events:
    time, event_type = event
    current_customers += event_type
    customer_counts.append((time, max(current_customers, 0)))

# Plotting
plt.step([event[0] for event in customer_counts], [event[1] for event in customer_counts], where='post')
plt.xlabel('Time')
plt.ylabel('Number of customers in the system')
plt.title('M/M/ Markov Jump Process (Queue)')
plt.show()
```

1. **Variability in System Size**: The plot shows fluctuations in the number of customers in the system over time. These fluctuations indicate the dynamic nature of the system, with varying levels of congestion or activity.

2. **Unbounded Queue Size**: Unlike the M/M/1 queue, which has a finite capacity, the M/M/ queue allows an infinite number of customers to be served simultaneously. This is reflected in the graph by the presence of higher peaks, indicating that the queue size can grow without bound during periods of high arrival rates.

3. **Steady-State Behavior**: Despite the fluctuations, the system appears to reach a steady-state behavior after some time. This suggests that the queue stabilizes over time, with the number of customers in the system exhibiting consistent patterns.

4. **System Dynamics**: The graph provides insights into the dynamics of the queueing system, including the arrival and departure rates of customers and how these factors influence the queue length over time.

5. **Queue Management**: Analyzing the fluctuations in the queue length can inform queue management strategies, such as adjusting service rates or implementing buffering mechanisms to handle variations in demand more effectively.

Overall, the graph provides valuable information about the behavior of the M/M/ Markov jump process, highlighting its dynamic nature and the variability in queue size over time.

# M/M/1 Process-Single Server Exponential Queueing System having Infinite Capacity- Main Process under Consideration
This is again a Markov Jump process and the only difference being that the number of servers is now limited to 1. This means that the customers arrive in the system as a Poisson Process with intensity $\lambda$ and if the server is busy then they wait in the queue until the server is free. Then the customer who has waited in the queue for the longest time gets the service for an $\text{Exp}(\mu)$ amount of time and the process continues until the queue is non-empty.

As regards departures, we can assume that departures are independent of arrivals in the following sense.
The server will take the same time with a customer whether the queue has 10 persons or 50 persons waiting.
So, let us stop the arrivals for a
moment, and think of $n$ persons in the queue. The queue will eventually empty as each person is served
in turn.

The probability of the queue reducing from  $n$  to  $n - 1$  in a time  $h$ is simply the probability that the service completes in that time. Let  $T_s$  denote the service time, then:

$$
\mathbb{P}(T_s \leq h) = 1 - e^{-\mu h}
$$

and for a very short time  $h \rightarrow 0$

$$
\mathbb{P}(T_s \leq h) \approx \mu h + o(h) . \quad (a)
$$

That is, in the absence of arrivals:

$$
\mathbb{P}(N_h = n - 1 | N_0 = n) = \mu h + o(h) . \quad (b)
$$


We see that $\mu $ represents the rate of departures, in the same way that $\lambda$ is the rate of arrivals. I have said that these processes are taken to be independent.

Therefore to include both arrivals and departures in the complete jump-rate matrix, we proceed as follows. Suppose the state of the system is $ N_t = n $, then in a very short time $ h $ there can be at most one arrival and/or one departure. Consider the jump rate for the queue to increase on this time. This means having exactly one arrival and no departures:

$$
q_{n,n+1} = \lim_{{h \to 0}} \frac{1}{h} \left( P(N_h = n + 1 | N_0 = n) \right) = \lim_{{h \to 0}} \frac{1}{h} \left( (\lambda h + o(h)) \times (1 - \mu h + o(h)) \right) . \quad 
$$

That is 

$$
q_{n,n+1} = \lambda . \quad 
$$

The jump rate for the queue to shorten in this time, $ h $, means having exactly one departure and no arrivals:

$$
q_{n,n-1} = \lim_{{h \to 0}} \frac{1}{h} \left( \mathbb{P}(N_h = n - 1 | N_0 = n) \right) = \lim_{{h \to 0}} \frac{1}{h} \left( (\mu h + o(h)) \times (1 - \lambda h + o(h)) \right) . \quad
$$

That is, 

$$
q_{n,n-1} = \mu . \quad 
$$

Finally, for the queue to stay the same length $ n \rightarrow n $, over this short time $ h $, requires either no arrival and no departures, or exactly one arrival and one departure.

$$
q_{n,n} = \lim_{h \to 0} \frac{1}{h} \left( \mathbb{P}(N_h = n|N_0 = n) - 1 \right) . \quad 
$$

that is,

$$
q_{n,n} = \lim_{h \to 0} \frac{1}{h} \left( (1 - \lambda h + o(h))(1 - \mu h + o(h)) + (\lambda h + o(h))(\mu h + o(h)) - 1 \right) . 
$$

Then neglecting all the tiny terms, that is using the fact that $ \lim_{{h \to 0}} \frac{o(h)}{h} = 0 $, we get:

$$
q_{n,n} = -\lambda - \mu . \quad 
$$

There is a special case when $ n = 0 $, that is when the queue is empty. In that case no departures are possible and: $ q_{0,0} = -\lambda $.

Thus, to summarize we can think of it intuitively as follows,

Let $N_{t}$ be the number of individuals in the system at time $t$. Thus, $N_{t}=0$ if the server is free, $N_{t}=1$ if one individual is receiving the service but the queue is empty and $N_{t}=x$ if one individual is receiving the service and the queue is non-empty with $x-1$ individuals waiting for their turn. The arrival rate $q_{i,i+1}=\lambda$ in this case as well. But for $i\geq 1$, the departure rate $q_{i,i-1}=\mu$ which is different from the previous case. This is mainly due to the fact that the next service can start only after the current individual(only 1 in this case) leaves, i.e, after an $\text{Exp}(\mu)$ amount of time. Thus, $q_i=-q_{ii}=\lambda+\mu$ for $i\geq 1$ and $q_0=-q_{00}=\lambda$. 
Hence the generator matrix $Q$ in this case is $$Q = \begin{pmatrix}
-\lambda & \lambda & 0 & 0 & \cdots \\
\mu & -(\lambda + \mu) & \lambda & 0 & \cdots \\
0 & \mu & -(\lambda + \mu) & \lambda & \cdots \\
0 & 0 & \mu & -(\lambda + \mu) & \cdots \\
\vdots & \vdots & \vdots & \vdots & \ddots \\
\end{pmatrix}
$$

The transition rate diagram can be represented as: 

We observe that $Q$ for the $M/M/1$ queue is a tridiagonal matrix in which all
elements above the main diagonal are equal and all elements below the main
diagonal are equal. With the exception of the topmost element, all elements of
the main diagonal are also equal and are negative sums. The topmost element of
the main diagonal is different because that state is a boundary state that has no
transition to a lower state. 

### Kolmogorov Equations

* **Forward**

We already know that by the Kolmogorov Forward equation, $\frac{dP_t}{dt}=P_t Q$, using the notation $P_t(x,y)=\mathbb{P}(N_t=y | N_0=x)$. Thus, by using these equations, we can make two cases.
* $y=0$,
$$
\begin{align*}
\frac{dP_t}{dt}(x,0)&=P_t Q(x,0)\\
&=\sum_y P_t(x,y)q_{y,0}\\&=P_t(x,0)q_{00}+P_t(x,1)q_{1,0}\\&=-\lambda P_t(x,0)+\mu P_t(x,1)
\end{align*}
$$

* $y>0,$

$$
\begin{align*}
\frac{dP_t}{dt}(x,y)&=P_t Q(x,y)\\
&=\sum_z P_t(x,z)q_{z,y}\\
&=P_t(x,y)q_{y,y}+P_t(x,y-1)q_{y-1,y}+P_t(x,y+1)q_{y+1,y}\\
&=-(\lambda+\mu) P_t(x,y)+\lambda P_t(x,y-1)+\mu P_t(x,y+1)
\end{align*}
$$


* **Backward**

Again, we already know that by the Kolmogorov Backward equation, $\frac{dP_t}{dt}=Q P_t $, using the notation $P_t(x,y)=\mathbb{P}(N_t=y | N_0=x)$. Thus, by using these equations, we can make two cases.
* $x=0$,

$$
\begin{align*}
\frac{dP_t}{dt}(0,y)&=Q P_t(0,y) \\
&=\sum_z q_{0,z}P_t(z,y)\\
&=q_{00}P_t(0,y)+q_{0,1}P_t(1,y)\\
&=-\lambda P_t(0,y)+\lambda P_t(1,y)
\end{align*}
$$

* $x>0,$

$$
\begin{align*}
\frac{dP_t}{dt}(x,y)&=QP_t (x,y)\\
&=\sum_z q_{x,z}P_t(z,y)\\
&=q_{x,x}P_t(x,y)+q_{x,x+1}P_t(x+1,y)+q_{x,x-1}P_t(x-1,y)\\
&=-(\lambda+\mu) P_t(x,y)+\lambda P_t(x+1,y)+\mu P_t(x-1,y)
\end{align*}
$$


****Theoretical Result 1.3** We admit that, for every $ i \in \mathbb{N} $,

$$
\frac{d}{dt} \mathbb{E}[N_t | N_0 = i] = \sum_{j \in \mathbb{N}} j \frac{dP_t}{dt}(i, j)
$$

Then,

$$
\frac{d}{dt} \mathbb{E}[N_t | N_0 = i] = \lambda - \mu \mathbb{P}(N_t \geq 1).
$$

*Proof.* The proof of this can be easily seen by using the Kolmogorov Forward equations.
Using the fact that 

$$ P_t f(i) = \mathbb{E} (f(N_t) | N_0 = i) \text{ for } i \in E, t \geq 0 \text{ and } f \text{ bounded,} $$

we get $$\mathbb{E}[N_t | N_0 = i] =P_t h(i)$$ 

where $h$ is the identity function.
Also, by the definition of $ P_t f(i)$, we have

$$ P_t h(i)=\sum_{j \in \mathbb{N}} P_t(i, j) h(j)= \sum_{j \in \mathbb{N}} P_t(i, j) j$$. 

On differentiating both sides w.r.t. $t$ we get the given condition.

Now, using the given condition and Kolmogorov Forward Equation, we get, 

$$\frac{d}{dt} \mathbb{E}[N_t | N_0 = i]=\sum_{j \in \mathbb{N}} j \frac{dP_t}{dt}(i, j) = 
\sum_{j \in \mathbb{N}}j\left(-(\lambda+\mu) P_t(i,j)+\lambda P_t(i,j-1)+\mu P_t(i,j+1)\right) $$

Now, 

$$\sum_{j=1}^{\infty}j P_t(i,j-1)=\sum_{j=0}^{\infty}(j+1) P_t(i,j)=\mathbb{E}[N_t+1 | N_0 = i]$$

Similarly, 

$$\sum_{j=1}^{\infty}j P_t(i,j+1)=\sum_{j=2}^{\infty}(j-1) P_t(i,j)=\sum_{j=2}^{\infty}j P_t(i,j)-\sum_{j=2}^{\infty} P_t(i,j)=\mathbb{E}[N_t | N_0 = i]-P_t(i,1)-\sum_{j=2}^{\infty} P_t(i,j)=\mathbb{E}[N_t | N_0 = i]-\sum_{j=1}^{\infty} P_t(i,j)=\mathbb{E}[N_t | N_0 = i]-\mathbb{P}(N_t \geq 1)$$

Now, substituting the above in the given condition, we get,

$$\frac{d}{dt} \mathbb{E}[N_t | N_0 = i]=\lambda(\mathbb{E}[N_t+1 | N_0 = i]-\mathbb{E}[N_t | N_0 = i])-\mu \mathbb{P}(N_t \geq 1)=\lambda - \mu \mathbb{P}(N_t \geq 1)$$ which is the desired result.

### Simulation for validating Theoretical Result 1.3 for i=0

#### Algorithm

1. **Define Function `mm1der`:**
   - Input: Arrival rate (`l`), service rate (`u`), time horizon (`Tim`)
   - Output: Times (`times`), states (`states`), derivative (`derivative`)
   - Initialize parameters:
     - `lambda_arrival` as `l`
     - `lambda_service` as `u`
     - `T` as `Tim`
   - Initialize state (`state`) as 0
   - Initialize empty lists `states` and `times`
   - Run a simulation loop until time `T`:
     - Generate time until the next event (`t`) using exponential distribution
     - Append `t` to `times`
     - Decide whether it's an arrival or a departure event based on the arrival and service rates
     - Update the state accordingly and ensure it remains non-negative
     - Append the updated state to `states`
   - Calculate the derivative using the finite difference method:
     - Set a small time increment `dt`
     - Find the state at time `1 + dt` and `1` in the `states` list
     - Calculate the derivative as the difference between these two states divided by `dt`
   - Return `times`, `states`, and `derivative`

2. **Define Function `LLN_Prob_Model`:**
   - Input: Number of simulations (`M`), arrival rate (`lamb`), service rate (`mu`), time horizon (`T`)
   - Output: Mean and standard deviation of the sample
   - Initialize an empty list `sample_new`
   - Run a loop for `M` simulations:
     - Simulate an M/M/1 queue for time `T` and get the state trajectory
     - Check if the final state is greater than or equal to 1
     - Append 1 for success or 0 for failure to `sample_new`
   - Calculate the mean and standard deviation of the outcomes
   - Return the mean and standard deviation

3. **Main Program:**
   - Set parameters: `vect_T`, `lamb`, `mu`, `r`, `M`
   - Call `LLN_Prob_Model` to get the probability (`Prob`) and standard deviation (`std`)
   - Calculate the expected derivative (`deriv_exp`) using function `mm1der`
   - Calculate the derivative of the calculated value (`deriv_calc`) using the formula provided
   - Compute the confidence interval (`CI`) for the probability using the asymptotic approach
   - Print the probability, expected derivative, calculated derivative, confidence interval, and relative error

This algorithm outlines the step-by-step process of simulating an M/M/1 queue, calculating the probability, and evaluating the derivatives as described in the provided code.
Here, we can explain what the variables are doing:

1. **Function `mm1der`:**

   - `lamb`: Represents the arrival rate of customers to the system.
   - `mu`: Represents the service rate of the system.
   - `T`: Represents the time horizon or the duration for which each simulation will run.

   - `std`: Represents the standard deviation of the outcomes.

3. **Main Program:**
   - `vect_T`: Specifies the time horizon for which to evaluate the probability and derivatives.
   - `r`: Represents the traffic intensity, calculated as the ratio of arrival rate to service rate (`lamb / mu`).
   - `deriv_exp`: Stores the expected derivative of the expected number of customers in the system.(LHS)
   - `deriv_calc`: Stores the calculated derivative of the expected number of customers in the system.(RHS)
   - `CI`: Represents the confidence interval for the probability of having at least one customer in the system at time `T`.
   - `mean`: Represents the mean probability calculated from the simulations.
   - `lenght_IC`: Represents the length of the confidence interval.
   - `relative error`: Indicates the relative error in the calculated confidence interval compared to the mean probability.

```python
def mm1der(l,u,Tim):
    # Parameters
    lambda_arrival = l  # Arrival rate
    lambda_service = u  # Service rate
    T = Tim              # Time horizon

    # Initial state
    state = 0  # Start with zero customers in the system

    # Simulate M/M/1 queue
    t = 0
    states = [state]
    times = [t]
    while t < T:
        # Generate time until the next event
        t += np.random.exponential(1 / (lambda_arrival + lambda_service))
        times.append(t)
    
        # Decide whether it's an arrival or a departure event
        p_arrival = lambda_arrival / (lambda_arrival + lambda_service)
        if np.random.rand() < p_arrival:
            state += 1
        else:
            state -= 1
    
        # Ensure the state remains non-negative
        state = max(state, 0)
    
        # Record the state
        states.append(state)
    
     # Calculate the derivative using finite difference method
    dt = 0.01  # small time increment
    expected_customers_next = states[np.searchsorted(times, 1 + dt)]
    expected_customers = states[np.searchsorted(times, 1)]
    derivative = (expected_customers_next - expected_customers) / dt
    
    
    return times, states, derivative

# Using Monte Carlo Simulation to find out the P(N_t>=1)
def LLN_Prob_Model(M, lamb, mu, T):
    sample_new = []
    for i in range(1, M):
        # Simulate an M/M/1 queue for time T and get the state trajectory(Algorithm for mm1() mentioned in the later simulation)
        N_t = mm1der(lamb, mu, T)[1]
        # Check if the final state is greater than or equal to 1 
        if N_t[-1] >= 1:
            sample_new.append(1)  # Success, append 1
        else:
            sample_new.append(0)  # Failure, append 0
    return np.mean(sample_new),np.std(sample_new)  # Calculate the mean of the outcomes and the standard deviation

# Parameters
vect_T = 100
lamb = 0.6
mu = 0.8
r = lamb / mu  # Traffic intensity(rho)
M = 10000  # Number of simulations

Prob,std=LLN_Prob_Model(M, lamb, mu, vect_T)
print("The Probability that N_t >= 1 is",Prob)
deriv_calc=lamb-(mu*Prob)
deriv_exp=mm1der(lamb,mu,vect_T)[2]

print("The Derivative of the Expected Value of N_t when starting from any fixed i:",deriv_exp)
print("The Calculated Value of the derivative when starting from any fixed i:",deriv_calc)

#Computation of the CI
mean=Prob
lenght_IC=2*1.96*std/np.sqrt(M)

print("Monte Carlo Probability,Asymptotic Confidence interval 95%")
print
print("CI={:1.10f} +/- {:1.6f} \n".format(mean,lenght_IC/2))
print("CI=[{:1.10f}, {:1.6f}] \n".format(mean-1.96*std/np.sqrt(M),mean+1.96*std/np.sqrt(M)))
print("Relative error = lenght_CI/MonteCarloProba={:1.10f} % \n".format(100*lenght_IC/mean))
print("Approximate Error in the Derivative:",np.abs(deriv_calc-deriv_exp)*100,"%")
```

### Stationary Distribution
Let us again look at the stationary distribution of this process to analyze its long term behaviour. Note that for this continuous time jump process $N_{t}$, the discrete time jump chain $X_{n}$ is nothing but a simple random walk with up probability $p=\frac{\lambda}{\lambda+\mu}$ down probability $q=\frac{\mu}{\lambda+\mu}$ and a reflecting barrier at 0. We already know that this Markov Chain is transient for $p>q$, positive recurrent for $p<q$ and null recurrent for $p=q$. Thus, we have the following conclusions:
* For $\lambda >\mu$, the queue is transient and thus eventually the length of the queue will grow longer and longer and never clear.
* For $\lambda =\mu$, the queue is null recurrent, so the queue will eventually clear but the expected time to do so would be infinite, so the queue will get very long and individuals may have to wait for a very long time to get their turn.
* For $\lambda<\mu$, the queue is positive recurrent, so again the individuals will have to wait for some time but that time won't be much large since the queue will clear out at fair regular intervals.

Note that it is quite intuitive that there would
be no limiting probabilities if $\lambda>\mu$. Let us suppose that is the case. Since individuals are arriving at a Poisson rate $\lambda$  it follows that the expected total number of arrivals by time $t$ is $\lambda t$.
On the other hand, what is the expected number of customers served by time $t$? If
there were always customers present, then the number of customers served would be
a Poisson process having rate $\mu$ since the time between successive services would be
independent exponentials having mean $\frac{1}{\mu}$. Hence, the expected number of customers
served by time $t$ is no greater than $\mu t$; and, therefore, the expected number in the system
at time $t$ is at least $$\lambda t-\mu t=(\lambda-\mu)t$$ 

Now, if $\lambda>\mu$, then the preceding number goes to infinity as $t$ becomes large. That
is, $\frac{\lambda}{\mu}>1$, the queue size increases without limit and there will be no limiting
probabilities.
Out of the above three cases,thus only the last case is ideal for us since in the other two, the expected time for the queue to be cleared is infinity. So we will talk only about the case when $\rho<1$.
Let us now look at the stationary distribution $\pi$ which will solve $\pi Q=0$. Using that we get:

$$
\lambda \pi_{i-1} - (\lambda + \mu)\pi_i + \mu\pi_{i+1} = 0     \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;        (3)$$
for $i\geq 1$ and $$
-\lambda \pi_0 + \mu \pi_1 = 0                                        \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(4)                          
$$

using the first column of $Q$. We also use the normalising condition in this case i.e, 

$$\sum_i\pi_i=1\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;(5)$$

These three equations can be solved by numerous approaches. Two of them are mentioned below.

#### Direct Approach
The equations (3) are a second order recurrence relations with constant coefficients. Its general solution is of the form,

$$\pi_i=c_1x_{1}^{i}+c_2x_{2}^{i}$$ 

where $x_1$ and $x_2$ are the roots of the quadratic equation,

$$\lambda-(\lambda+\mu)x+\mu x^2=0$$

This equation has two roots, namely $x=1$ and $x=\frac{\lambda}{\mu}=\rho$. So all the solutions to (3) are of the form

$$\pi_i=c_1+c_2\rho^{i}$$

Now, using (5), we easily get that $c_1=0$ and again using (5) we get $c_2=1-\rho$. Thus, finally, 

$$\pi_i=(1-\rho)\rho^i$$

#### Recursive Approach

One can use (4) to express  $\pi_1$ in $\pi_0$ yielding

$$ \pi_1 = \rho \pi_0 $$

Substitution of this relation into (3) for $ i = 1 $ gives

$$ \pi_2 = \rho^2\pi_0 $$

By substituting the relations above into (3) for $ i = 2 $ we obtain  $\pi_3$ , and so on. Hence we can recursively express all probabilities in terms of $\pi_0 $, yielding

$$ \pi_i = \rho^n \pi_0 , \quad i = 0, 1, 2, \ldots. $$

The probability $ \pi_0 $ finally follows from the normalization equation (5).

Thus, we have a statitionary distribution $\pi_i=(1-\rho)\rho^i$ which is a geometric distribution $\text{Geom}(1-\rho)$.

**Theorem 1.6** Let $ (N_{t})$ be an $M/M/1$ process with arrivals at rate $\lambda$ and service at rate $\mu>\lambda$. The process has the stationary distribution $\pi_i=(1-\rho)\rho^i$, corresponding to $N_{t} \sim Geom(1-\rho)$, where $\rho = \frac{\lambda}{\mu}<1 $.

*Proof.* The proof goes along the same lines as Theorem 1.2. We know that this $ \pi $ is a distribution. We must check that $\pi $ fulfills $ \pi Q = 0 $. We've already checked (4). For $i \geq 1 $, we have from (3),

$$
\begin{align*}
\lambda \pi_{i-1} - (\lambda + \mu)\pi_i + \mu\pi_{i+1} &= \lambda (1-\rho)\rho^{i-1} - (\lambda + \mu) (1-\rho)\rho^{i} + \mu(1-\rho)\rho^{i+1} \\
&= (1-\rho)\rho^{i-1}(\lambda+\rho^2\mu)-(\lambda + \mu) (1-\rho)\rho^{i}\\
&= (1-\rho)\rho^{i}(\frac{\lambda}{\rho}+\rho\mu)-(\lambda + \mu) (1-\rho)\rho^{i}\\
&= (\lambda + \mu) (1-\rho)\rho^{i}-(\lambda + \mu) (1-\rho)\rho^{i}\\
&= 0,
\end{align*}
$$


Again, we see that this process indeed has a stationary distribution and simultaneously it is irreducible. Thus, it must be positive recurrent and $\pi$ is also the limiting or equilibrium distribution. So we can use the ergodic theorem to say that the long-run amount of time spent in each state is governed by $\pi$. 

### Some Key Features of the Process

**Theoretical Result 1.4** The long run average number of individuals in the process is $$\mathbb{E}(Geom(1-\rho))=\sum_ii \pi_i=\sum_ii (1-\rho)\rho^i=\frac{\rho}{1-\rho}=\frac{\lambda}{\mu-\lambda}=L_S$$

**Theoretical Result 1.5** The average time spent in the system is $W_S=\frac{\frac{1}{\mu}}{1-\rho}$.

*Proof:* Using the PASTA property, we know that the average number of customers in the system seen by an arriving customer equals to $L_S$ and each of them including the one being serviced has a residual service time $\frac{1}{\mu}$. The particular individual also has to wait for its own service time($=\mathbb{E}(Exp(\mu))=\frac{1}{\mu}$). Thus, $$W_S=L_S\frac{1}{\mu}+\frac{1}{\mu}$$. Also, by Little's law, we have the following relation, $$L_S=\lambda W_S$$. On solving the above two equations, we get the desired result.

**PASTA(Poisson Arrivals See Time Averages) Property** The PASTA property refers to the expected state of a queueing system as seen by an arrival from a Poison process. An arrival from a Poisson process observes the system as if it was arriving at a random moment in time. Therefore, the expected value of any parameter of the queue at the instant of a Poisson arrival is
simply the long-run average value of that parameter.

*Proof.*  Consider a period of length $ t $, divided into three non-overlapping regions with lengths $ a_1 $, $ b $, and $ a_2 $. We'll show that the probability that a Poisson process produces one arrival in the period of length $ b $ is the same as the probability of a randomly chosen point being in the interval $ b $. This argument can be extended to a general case with any number of arrivals.

The probability that a single Poisson arrival on $ [0, t] $ occurs in the interval of length $ b $ can be written as

$$
\frac{P\{0 \text{ arrivals in } a_1 \text{ and 1 arrival in } b \text{ and 0 arrivals in } a_2\}}{P\{1 \text{ arrival in } t \}}
$$

The denominator comes from conditioning on the probability of getting only one arrival during the entire period of length $ t $. Split the numerator into its three independent components, then use the Poisson distribution to calculate the probability of getting 0 or 1 arrivals in each period.

$$
\frac{e^{-\lambda a_1} \cdot \lambda b e^{-\lambda b} \cdot e^{-\lambda a_2}}{\lambda t e^{-\lambda t}}
$$

Adding the exponents, we obtain

$$
\frac{\lambda b e^{-\lambda(a_1+b+a_2)}}{\lambda t e^{-\lambda t}}
$$

But $ a_1 + b + a_2 = t $, so we can cancel the exponential terms and $ \lambda $ to give

$$
P(\text{one arrival in period of length } b) = \frac{b}{t}
$$

This is exactly the probability that a randomly chosen point in $ [0, t] $ lands in the interval of length $ b $. Thus, arriving from a Poisson process is statistically indistinguishable from arriving at a random moment in time.

**Little's Law** The time average number of customers in a queueing
system, $L_S$, is equal to the rate at which customers arrive and enter the system, $\lambda$,  the average
sojourn time of a customer, $W_S$, i.e, $$L_S=\lambda W_S$$
The proof for this law is skipped.

#### Conditional Distribution w.r.t. $W$
Let us suppose a given customer spends a total of $t$ time units in the system which has been operational for a long time i.e., it is in steady state. Let us focus on the conditional distribution of $N$, the number of other customers that were present in the system when the customer arrived and let $W$ be the amount of time a customer spends in the system, we look at $\mathbb{P}(N=n | W=t)$,

$$
\mathbb{P}(N = n | W = t) = \frac{f_{N,W}(n, t)}{f_{W}(t)} = \frac{\mathbb{P}(N = n)f_{W|N}(t|n)}{f_{W}(t)}
$$

where $f_{W|N}(t|n)$ is the conditional density of $W$ given that $N=n$ and $f_{W}(t)$ is the density of $W$. Given that $N=n$, the time that the customer spends in the system is distributed as a sum of $n+1$ i.i.d. exponential random variables with a common parameter $\mu$. Thus, the conditional distribution of $W$ given $N=n$ is the gamma distribution with parameters $n+1$ and $\mu$. Therefore, 

$$
\mathbb{P}(N = n|W = t) = B \mathbb{P}(N = n)\mu\frac{e^{-\mu t}(\mu t)^{n}}{n!}
$$

$$
= B\left(\frac{\lambda}{\mu}\right)^{n} \left(1 - \frac{\lambda}{\mu}\right)\mu e^{-\mu t} \frac{(\mu t)^{n}}{n!} \quad \text{(by PASTA)}
$$

$$
= A\frac{(\lambda t)^n}{n!}
$$

where $A=B\mu e^{-\mu t}\left(1 - \frac{\lambda}{\mu}\right)$, $B =\frac{1}{f_{W}(t)}$ and $A$ is independent of $n$. If we sum over all $n$ we get,

$$1 = \sum_{n=0}^{\infty} \mathbb{P}(N=n|W=t) = A \sum_{n=0}^{\infty} \frac{(\lambda t)^{n}}{n!} = Ae^{\lambda t}$$

which gives us the value of $A=e^{-\lambda t}$. Thus, $$\mathbb{P}(N = n|W = t) = e^{-\lambda t}\frac{(\lambda t)^{n}}{n!}$$

Therefore, the conditional distribution of the number seen by an arrival who spends a
total of $t$ time units in the system is the Poisson distribution with mean $\lambda t$.

**Remark 1.1** The amount of time a customer spends in the system $W\sim\text{Exp}(\mu-\lambda)$.

*Proof.* Now, we have already defined $$f_{W}(t)=\frac{1}{B} = \frac{1}{A}\mu(1 - \lambda/\mu)e^{-\mu t} = (\mu - \lambda)e^{-(\mu - \lambda)t}$$

#### Algorithm 1

Mentioned below is an algorithm to simulate an M/M/1 process with explanations for the provided code:

1. **Import Libraries**: Import the required libraries for numerical computation and plotting. In this case, NumPy for numerical computation and Matplotlib for plotting and define a function which will return the states vector and the times vector.

2. **Define Parameters**: Set the parameters for the M/M/1 queue simulation. This includes the arrival rate (`lambda_arrival`=$\lambda$), service rate (`lambda_service`=$\mu$), and the time horizon (`T`), which defines the duration of the simulation.

3. **Initialize State**: Set the initial state of the system. In this case, initialize the number of customers in the system to zero.

4. **Simulate M/M/1 Queue**:
   - Initialize time (`t`) to zero.
   - Initialize lists to store the system states (`states`) and corresponding times (`times`). The initial state and time are appended to these lists.
   - Start a loop that runs until the simulation time reaches the defined time horizon (`T`).
   - Inside the loop:
     - Generate the time until the next event (either arrival or departure) using an exponential distribution with the rate parameter equal to the sum of arrival and service rates($\sim\text{Exp}(\lambda+\mu)$).
     - Increment the simulation time by the generated time and append it to the `times` list.
     - Decide whether the event is an arrival or a departure based on a random number generated from a uniform distribution. The probability of an arrival is calculated based on the arrival and service rates.
     - Update the system state accordingly by incrementing or decrementing the number of customers in the system.
     - Ensure that the system state remains non-negative by setting it to zero if it becomes negative.
     - Append the updated state to the `states` list.
    -Return the `times` and `states` vectors
     
5. **Plot the M/M/1 Queue**:
   - Use Matplotlib to plot the number of customers in the system (`states`) against time (`times`) using a step plot (`plt.step`).
   - Label the axes appropriately and give the plot a title.

This algorithm simulates the $M/M/1$ queue process by generating random events for arrivals and departures based on exponential distributions and updating the system state accordingly. Finally, it visualizes the evolution of the queue over time using a step plot.

```python
import numpy as np
import matplotlib.pyplot as plt
def mm1(l,u,Tim):
    # Parameters
    lambda_arrival = l  # Arrival rate
    lambda_service = u  # Service rate
    T = Tim              # Time horizon

    # Initial state
    state = 0  # Start with zero customers in the system

    # Simulate M/M/1 queue
    t = 0
    states = [state]
    times = [t]
    while t < T:
        # Generate time until the next event
        t += np.random.exponential(1 / (lambda_arrival + lambda_service))
        times.append(t)
    
        # Decide whether it's an arrival or a departure event
        p_arrival = lambda_arrival / (lambda_arrival + lambda_service)
        if np.random.rand() < p_arrival:
            state += 1
        else:
            state -= 1
    
        # Ensure the state remains non-negative
        state = max(state, 0)
    
        # Record the state
        states.append(state)
    return times, states
# Plot the M/M/1 queue
times,states=mm1(0.6,0.8,100)
plt.step(times, states, where='post')
plt.xlabel('Time')
plt.ylabel('Number of customers in the system')
plt.title('M/M/1 Markov Jump Process (Queue)')
plt.show()
```

#### Algorithm 2 (Event-Driven Approach)

Another way to simulate the M/M/1 queue is by using an event-driven simulation. In this approach, instead of advancing time continuously and checking for arrivals and departures at each time step, we keep track of the next arrival and departure events and simulate them one by one until the simulation time reaches the desired horizon. This method can be more efficient, especially for long simulation times, as it only processes events when they occur.
Here's a detailed algorithm for the event-driven simulation of an M/M/1 queue:

1. **Initialize Parameters**:
   - Set the arrival rate `lambda_arrival`, service rate `lambda_service`, and time horizon `T`.

2. **Initialize Simulation**:
   - Set the initial state of the system `state` to 0 (no customers in the system).
   - Set the initial time `t` to 0.

3. **Initialize Event List**:
   - Create an empty list `events` to store arrival and departure events.
   - Generate the first arrival event with an exponential inter-arrival time sampled from the distribution with rate `lambda_arrival`.

4. **Simulation Loop**:
   - While the current time `t` is less than the time horizon `T`, repeat the following steps:
     - Get the next event time `dt` and event type `event_type` from the first event in the `events` list.
     - Update the current time `t` by adding the event time `dt`.
     - If the event type is an "arrival":
       - Increase the system state `state` by 1 to represent the arrival of a customer.
       - Generate the next arrival event and append it to the `events` list with an exponential inter-arrival time.
       - If the system was idle (i.e., `state` was 0) before this arrival, schedule a departure event by generating it with an exponential service time and adding it to the `events` list.
     - If the event type is a "departure":
       - Decrease the system state `state` by 1 to represent the departure of a customer.
       - If there are still customers in the system (i.e., `state` > 0), schedule the next departure event by generating it with an exponential service time and adding it to the `events` list.
     - Record the current time `t` and system state `state` in the `times` and `states` lists, respectively.
     - Remove the processed event from the `events` list.
     - Sort the `events` list by event time to ensure the next event is the earliest one.

5. **Plotting**:
   - Plot the system state over time using the `times` and `states` lists.
   - Label the x-axis as "Time" and the y-axis as "Number of customers in the system".
   - Provide a title for the plot, such as "M/M/1 Markov Jump Process (Queue) - Event-driven Simulation".

6. **Display the Plot**:
   - Show the generated plot.

7. **End**.

This algorithm outlines the steps involved in simulating an M/M/1 queue using an event-driven approach, which efficiently handles both arrival and departure events while ensuring correctness and accuracy in the simulation.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lambda_arrival = 0.6  # Arrival rate
lambda_service = 0.8  # Service rate
T = 10000              # Time horizon

# Initialize simulation
state = 0  # Start with zero customers in the system
t = 0      # Initial time

# Event list containing arrival and departure events
events = [(np.random.exponential(1 / lambda_arrival), 'arrival')]

# Lists to store event times and corresponding system states
times = [t]
states = [state]
D = []  # List to store times between successive departures(useful for the analysis of the departure process)

# Event-driven simulation
while t < T:
    # Get the next event time and type
    dt, event_type = events[0]
    t += dt
    
    # Update the system state based on the event type
    if event_type == 'arrival':
        state += 1
        events.append((np.random.exponential(1 / lambda_arrival), 'arrival'))
        if state == 1:  # If the system was idle, schedule a departure
            events.append((np.random.exponential(1 / lambda_service), 'departure'))
    else:  # departure
        state -= 1
        D.append(dt) # Record time between successive departures
        if state > 0:  # If there are still customers, schedule the next departure
            events.append((np.random.exponential(1 / lambda_service), 'departure'))
    
    # Record the event time and system state
    times.append(t)
    states.append(state)
    
    # Remove the processed event
    events.pop(0)
    
    # Sort the event list by event time
    events.sort()

# Plot the M/M/1 queue
plt.step(times, states, where='post')
plt.xlabel('Time')
plt.ylabel('Number of customers in the system')
plt.title('M/M/1 Markov Jump Process (Queue) - Event-driven Simulation')
k=len(D)
print(len(D))
plt.show()
```

### Departure Process
Define $D_0 = 0$ and successive departure times:

$$
D_{n+1} = \inf \{ t > D_n : N_t - N_{t-1} = -1 \} \text{ for } n \geq 0.
$$



Let us study the process $ V_n = X_{D_n}, n \geq 0 $, i.e. the process of queue lengths after departures. By the memoryless property of $ \text{Exp}(\lambda) $, the geometric random variables $ X_n, n \geq 1 $, that record the number of new customers between $ D_{n-1} $ and $ D_n $, are independent. Therefore, $ (V_n)_{n \geq 0} $ is a Markov chain, with transition probabilities

$$
d_{k,k-1+m} = \begin{cases}
\frac{\lambda}{\lambda + \mu} & \text{for } k \geq 1, m \geq 0 \\
\frac{\mu}{\lambda + \mu} & \text{for } k = 0, m \geq 1
\end{cases}
$$

For $ k = 0 $, we get $ d_{0,m} = d_{1,m} \geq 0 $, since the next service only begins when a new customer enters the system.

**Proposition 1.1** $ V $ also has stationary distribution $ \pi $.

Proof: A simple calculation shows that with $ \rho = \lambda/\mu $ and $ q = \lambda/(\lambda + \mu) $
$$
\sum_{k \in \mathbb{N}} \pi_{k}d_{k,n} = \pi_{0}d_{0,n} + \sum_{k=1}^{n+1} \pi_{k}d_{k,n} = (1 - \rho) q^n + (1 - \rho)(1 - q)q^{n+1} \sum_{k=1}^{n+1} \left(\frac{\rho}{q}\right)^k = \pi_n,
$$
after bringing the partial geometric progression into closed form and appropriate cancellations.



Note that the conditional distribution of $ D_{n+1} - D_n $ given $ V_n = k $ is the distribution of a typical service time $ G \sim \text{Exp}(\mu) $ if $ k \geq 1 $ and the distribution of $ Y + G $, where $ Y \sim \text{Exp}(\lambda) $ is a typical interarrival time, if $ k = 0 $ since we have to wait for a new customer and his service. We can also calculate the unconditional distribution of $ D_{n+1} - D_n $, at least if $ V $ is in equilibrium.



****Theoretical Result 1.9** Let $D$ denote the time between successive departures in a stationary $M/M/1$
queue with $\lambda<\mu$. Then $D\sim \text{Exp}(\lambda).$

*Proof.* 
Let us first state a simple lemma that we are going to use:

*Lemma 1.1* If $X_1, X_2$ are iid exponential random variables with parameter $\lambda_1$ and $\lambda_2$ respectively, then $Z=X_1+X_2$ is a probability density model with $f_Z(t)=\frac{\lambda_1 \lambda_2}{\lambda_2-\lambda_1}(e^{-\lambda_1 t}-e^{-\lambda_2 t}).$

*Proof of Lemma.* We can use Convolution of random variables to prove this lemma.
$$
f(x_1) = 
\begin{cases} 
\lambda_1 e^{-\lambda_1 x_1}, & \lambda_1 > 0, x_1 \geq 0 \\
0, & \text{otherwise}
\end{cases}
$$

$$
\begin{align*}
\text{Let } Z &= X_1 + X_2, \text{ then } X_2 = Z - X_1, \\
\text{then, } f_{X_1, X_2}(x_1, z - x_1) &= \lambda_1 e^{-\lambda_1 x_1} \cdot \lambda_2 e^{-\lambda_2 (z - x_1)}\mathbb{1}_{(0, z)}(x_1)
\end{align*}
$$

$$
\begin{align*}
\text{By the concept of Convolution of random variables,} \\
f_Z(z) &= \int_{-\infty}^{\infty} f_{X_1, X_2}(x_1, z - x_1) \, dx_1 \\
&= \int_{0}^{z} \lambda_1 e^{-\lambda_1 x_1} \cdot \lambda_2 e^{-\lambda_2 (z - x_1)} \, dx_1 \\
&= \lambda_1 \lambda_2 e^{-\lambda_2 z} \int_{0}^{z} e^{(\lambda_2 - \lambda_1) x_1} \, dx_1 \\
&= \lambda_1 \lambda_2 e^{-\lambda_2 z} \left[ \frac{1}{\lambda_2 - \lambda_1} \left[ e^{(\lambda_2 - \lambda_1) x_1} \right]_0^z \right] \\
&= \frac{\lambda_1 \lambda_2}{\lambda_2 - \lambda_1} \left( e^{-\lambda_1 z} - e^{-\lambda_2 z} \right)
\end{align*}
$$

$$
\text{where } \lambda_1 > 0, \lambda_2 > 0, z \geq 0
$$

This is basically a Probabilty Model and thus we need to check its validity.

$$
\begin{align*}
\text{For the model } f_Z(z) \text{ to be a valid model, it suffices that} \\
\int_{-\infty}^{\infty} f_Z(z) \, dz &= 1 \\
&= \int_{0}^{\infty} \frac{\lambda_1 \lambda_2}{\lambda_2 - \lambda_1} \left( e^{-\lambda_1 z} - e^{-\lambda_2 z} \right) \, dz \\
&= \frac{\lambda_1 \lambda_2}{\lambda_2 - \lambda_1} \left[ \frac{1}{\lambda_1} \left( 1 - e^{-\lambda_1 z} \right) - \frac{1}{\lambda_2} \left( 1 - e^{-\lambda_2 z} \right) \right]_{0}^{\infty} \\
&= \frac{\lambda_1 \lambda_2}{\lambda_2 - \lambda_1} \left[ \frac{1}{\lambda_1} - \frac{1}{\lambda_2} \right] \\
&= 1
\end{align*}
$$

Similarly, we can find the Cumulative Density function for the same.

$$
\begin{align*}
\text{By definition, the cdf is derived by: } F(z) &= P(Z \leq z) \\
&= \int_{-\infty}^{z} f_Z(t) \, dt \\
&= \frac{\lambda_1 \lambda_2}{\lambda_2 - \lambda_1} \int_{0}^{z} \left( e^{-\lambda_1 t} - e^{-\lambda_2 t} \right) \, dt \\
F(z) &= 1 + \frac{\lambda_1}{\lambda_2 - \lambda_1} \left( e^{-\lambda_2 z} \right) - \frac{\lambda_2}{\lambda_2 - \lambda_1} \left( e^{-\lambda_1 z}\right)
\end{align*}
$$

Now keeping that lemma in mind, we can proceed with the proof as follows.
If a customer leaves the system busy, the time until the next departure is the time of
a service. If a customer leaves the system empty, the time until the next departure
is the time until an arrival plus the time of a service.

Let $A=\{\text{System is left empty}\}$

$B=\{\text{System is left busy}\}$

Then, $$\mathbb{P}(A)=1-\frac{\lambda}{\mu}$$ and $$\mathbb{P}(B)=\frac{\lambda}{\mu}$$


Also, $A$ and $B$ are disjoint sets and $\mathbb{P}(A \cup B)=1$, so if C is another set, then

$$\mathbb{P}(C)=\mathbb{P}(C \cap (A\cup B))=\mathbb{P}((C\cap A)\cup(C\cap B))=\mathbb{P}(C\cap A)+\mathbb{P}(C\cap B)=\mathbb{P}(C | A)\mathbb{P}(A)+\mathbb{P}(C | B)\mathbb{P}(B)$$

Thus,

$$
\begin{align*}
\mathbb{P}(D\leq t) &= \mathbb{P}(D\leq t | \text{System is left empty})\mathbb{P}(\text{System is left empty}) + \mathbb{P}(D\leq t | \text{System is left busy})\mathbb{P}(\text{System is left busy})\\
&=\mathbb{P}(X+Y\leq t)\left(1-\frac{\lambda}{\mu}\right)+\mathbb{P}(Y\leq t)\frac{\lambda}{\mu}\\
\end{align*}
$$

where $X$ has the distribution of interarrival times($\sim\text{Exp}(\lambda)$) and $Y$ has the distribution of service times($\sim\text{Exp}(\mu)$) and we have simultaneously used the Theoretical Result 1.5.

Now, simply using the Lemma and substituting the CDF of the model that we used and the CDF of an exponential distribution, we get,
$$
\begin{align*}
\mathbb{P}(D\leq t) &=\left( 1 + \frac{\lambda}{\mu - \lambda} \left( e^{-\mu t} \right) - \frac{\mu}{\mu - \lambda} \left( e^{-\lambda t}\right)\right)\left(1-\frac{\lambda}{\mu}\right)+(1-e^{-\mu t})\frac{\lambda}{\mu}\\
&=1-e^{-\lambda t}
\end{align*}
$$

which is the CDF of an exponential distribution with parameter $\lambda$. In fact it can also be shown that the interdeparture times are independent and thus the Output of an $ M/M/1$ system is also a Poisson Process of intensity $\lambda$.

**Corollary 1.1 (Burke's Theorem)** The Inter Departure Times of an $ M/M/1$ system are independent and the output of an $ M/M/1$ system is also a Poisson Process of intensity $\lambda$.

*Proof.* The
strong Markov property at $D_n$ makes the system start afresh conditionally independently
of the past given $V_n$. Since $D_1, \ldots, D_n  D_{n1}$ are independent of $V_n$, they are then also
independent of the whole post-$D_n$ process.

The argument is very subtle, because the post-$ D_n $ process is actually not independent of the whole pre-$ D_n $ process, just of the departure times. The result, however, is not surprising since we know that $ N $ is reversible, and the departure times of $ N $ are the arrival times of the time-reversed process, which form a Poisson process of rate $ \lambda $.

```python
import numpy as np
import matplotlib.pyplot as plt
def mm1(l,u,Tim):
    # Parameters
    lambda_arrival = l  # Arrival rate
    lambda_service = u  # Service rate
    T = Tim              # Time horizon

    # Initial state
    state = 0  # Start with zero customers in the system

    # Simulate M/M/1 queue
    t = 0
    states = [state]
    times = [t]
    while t < T:
        # Generate time until the next event
        t += np.random.exponential(1 / (lambda_arrival + lambda_service))
        times.append(t)
    
        # Decide whether it's an arrival or a departure event
        p_arrival = lambda_arrival / (lambda_arrival + lambda_service)
        if np.random.rand() < p_arrival:
            state += 1
        else:
            state -= 1
    
        # Ensure the state remains non-negative
        state = max(state, 0)
    
        # Record the state
        states.append(state)
    return times, states
# Plot the M/M/1 queue
times,states=mm1(0.6,0.8,1000)
plt.step(times, states, where='post')
plt.xlabel('Time')
plt.ylabel('Number of customers in the system')
plt.title('M/M/1 Markov Jump Process (Queue)')
plt.show()
```

Similarly, since we know the invariant distribution, thus, we can also validate the following theoretical result.

***Theoretical Result 1.10** $$\lim_{T\rightarrow \infty}\mathbb{P}(N_T=x)=\pi(x)$$ where $\pi$ is the invariant distribution of $N_t$.

*Proof.(Sketch)*
This is just a consequence of the Limit Theorem for Continuous Time Markov Chains where since we know that our MJP $N_t$ is irreducible and recurrent and on top of that it is also positive recurrent, thus there exists a unique stationary distribution $\pi$ and this is also the limiting distribution. We basically use the fact that

$$ \lim_{t \to \infty} P_t(x, y) = \lim_{t \to \infty} \frac{1}{t} \int_0^t P_s(x, y) \, ds $$. 

Now, the RHS represents nothing but the limiting proportion of time spent
in $y\in S$ when we start at $x\in S$. This proportion can also be written as 

$$ \frac{\mathbb{E} \left( \int_0^{\tau_x} \mathbb{1}_{N_s = y} \, ds \bigg|{X_0 = x}\right) , \quad y \in S}{ \mathbb{E} (\tau_{x} | X_0 = x)} $$, 

where $\tau_x$ represents the continuous hitting time and then the result follows from the limit theorem and thus this ratio is equal to $\pi(x)$




### Simulation of Theoretical  Result 1.10

#### Algorithm

1. **Initialization**:
   - Define the function `LLN_Prob_Model` which calculates the probability of the Markov Jump Process being in state `x` at time `T` using a large number of simulations.
   - Initialize parameters such as `M` (number of simulations), `lamb` (arrival rate), `mu` (service rate), `T` (time), and `x` (target state).

2. **Simulation Loop**:
   - Create an empty list `sample_new` to store the outcomes of each simulation.
   - Iterate `M` times:
     - Simulate an M/M/1 queue for time `T` using the `mm1` function, which returns the state trajectory `N_t` of the queue.
     - Check if the final state of the queue at time `T` is equal to the target state `x`.
     - If the final state matches the target state, append `1` to `sample_new`, indicating success. Otherwise, append `0`.
   - Calculate the mean of `sample_new`, representing the estimated probability of the queue being in state `x` at time `T`.

3. **Main Simulation Loop**:
   - Define parameters such as `vect_T` (list of time points to analyze), `lamb`, `mu`, `r` (traffic intensity), `M`, and `x_n` (list of states to consider).
   - Calculate the invariant probabilities `pi_x` for each state `x` using the M/M/1 queue parameters.
   - For each time `T` in `vect_T`:
     - Calculate the Monte Carlo probabilities for each state `x` using the `LLN_Prob_Model` function.
     - Plot the calculated Monte Carlo probabilities and the invariant probabilities for the same states.
     - Label the axes, add a title, legend, and grid to the plot.
     - Show the plot.

4. **Plotting**:
   - Create a new figure for each time `T` and display the plot.
   
5. **Conclusions**:
- The Monte Carlo probabilities converge to the theoretical stationary probabilities as the time (T) increases.
- The accuracy of the Monte Carlo estimation improves with an increasing number of simulations (M).
- The traffic intensity (r = /) affects the convergence rate of Monte Carlo probabilities to the stationary probabilities.
- Comparing Monte Carlo and theoretical probabilities provides insights into the performance of the M/M/1 queue model.
 
This algorithm iterates over different time points `T`, simulates the M/M/1 queue model, calculates probabilities, and visualizes the results to compare the Monte Carlo probabilities with the theoretical stationary probabilities.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate the probability of the Markov chain being in state x at time T
def LLN_Prob_Model(M, lamb, mu, T, x):
    sample_new = []
    for i in range(1, M):
        # Simulate an M/M/1 queue for time T and get the state trajectory
        N_t = mm1(lamb, mu, T)[1]
        # Check if the final state matches the target state x
        if N_t[-1] == x:
            sample_new.append(1)  # Success, append 1
        else:
            sample_new.append(0)  # Failure, append 0
    return np.mean(sample_new)  # Calculate the mean of the outcomes

# Parameters
vect_T = [5, 100, 2000]  # Time points to analyze
lamb = 0.6
mu = 0.8
r = lamb / mu  # Traffic intensity(rho)
M = 5000  # Number of simulations
x_n = np.arange(10)  # States to consider
pi_x = [(1 - r) * (r ** x) for x in x_n]  # Invariant probabilities

# Plotting
for T in vect_T:
    # Calculate Monte Carlo probabilities for each state x
    prob_vect = [LLN_Prob_Model(M, lamb, mu, T, x) for x in x_n]
    # Plot calculated Monte Carlo probabilities and invariant probabilities
    plt.plot(x_n, prob_vect, color="blue", label="Monte Carlo Probabilities")
    plt.plot(x_n, pi_x, color="red", label="Invariant Probabilities")
    # Add labels, title, legend, and grid
    plt.xlabel('States (x)')
    plt.ylabel('Probability P(N_T = x)')
    plt.title('Monte Carlo Probability vs Stationary Probability for T = {}'.format(T))
    plt.legend()
    plt.grid(True)
    plt.show()
```
